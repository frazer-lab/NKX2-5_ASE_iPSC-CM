```r
setwd("/home/paola/2017/anotations/")
library (stringr)
dir="/frazer01/projects/CARDIPS/analysis/family1070/ase_chips_v3/analysis"
```

```r
###### Closest Gene Annotation
gi=read.table("/frazer01/publicdata/gencode_v19_20151104/gene_info.tsv", header=T)
pro=c("/home/paola/BedFiles/NKXclosestGene.txt","/home/paola/BedFiles/SRFclosestGene.txt",
      "/home/paola/BedFiles/H3KclosestGene.txt","/home/paola/BedFiles/H3KIPSclosestGene.txt")

##### Motif disruption calculated here ####to be UPDATED######

motR=c('/home/paola/Motif_AnalysisR/20_bp/PV.05_NKX/Mix_mot/AlleleAlteringSites_scores_1.txt',
        '/home/paola/Motif_AnalysisR/20_bp/PV.05_SRF/Mix_mot/AlleleAlteringSites_scores_1.txt')

### Haploreg annotations

HAP=c("NKX_haploreg.tab", "SRF_haploreg.tab", "H3K_CM_haploreg.tab","H3K_iPS_haploreg.tab")
QTL=c("NKX_eQTLS.txt" ,"SRF_eQTLS.txt", "H3K_CM_eQTLS.txt", "H3K_iPS_eQTLS.txt")

#### Gwas annotations
gwasCAT<- read.csv('/home/paola/2017/anotations/GWAS/GwasForFinalAnnotation.csv', header=T)


##### left ventricle expression annot
lv<-read.csv('/home/paola/Annotations/GTEx Portal.csv', header=T)
aa<-read.csv('/home/paola/Annotations/GTEx Portal AA.csv', header=T)

lv_eqtls<-read.table("/publicdata/gtex_v6/Heart_Left_Ventricle_Analysis.snpgenes",  sep="\t", header=T)
lv_eqtls$snpID<-paste("chr",lv_eqtls$snp_chrom, ":", lv_eqtls$snp_pos, sep="")
lv<-merge(lv, lv_eqtls[,c('gene', 'snpID' )], by.x='Gencode.Id', by.y='gene')

aa_eqtls<-read.table("/publicdata/gtex_v6/Heart_Atrial_Appendage_Analysis.snpgenes",  sep="\t", header=T)
aa_eqtls$snpID<-paste("chr",aa_eqtls$snp_chrom, ":", aa_eqtls$snp_pos, sep="")
aa<-merge(aa, aa_eqtls[,c('gene', 'snpID' )], by.x='Gencode.Id', by.y='gene')

```

```r
nm=c("NKX", "SRF", "H3K_CM", "H3K_iPS")
m=data.frame()
for (i in 1:3){
  cat(nm[i], "\n")
  ase=read.csv(paste(dir, "/HaploregOverlap/SummaryAnnotations.", nm[i], ".csv", sep=""), header=T)
  ase=subset(ase, ase$combined_pv<0.005)
  
  
  hap=read.table(paste(dir, 'HaploregOverlap', HAP[i], sep="/"), header=F, sep="\t")
  colnames(hap)=c("chrom", "pos", "id", "ref", "alt", "cons", "motif", "gene","chromHMM", "DNAse")
  hap=hap[,c( "id", "motif", "gene","chromHMM", "DNAse")]
  ase=merge(ase,hap,by.x="snpID", by.y="id", all.x=T)
 
  qtl<-read.table(paste(dir,'GtexOverlap', QTL[i], sep="/"), fill=T, stringsAsFactors=F)
  colnames(qtl)<-c("snp","qtl")
  ase=merge(ase,qtl,by.x="snpID", by.y="snp", all.x=T)
  
  gene=read.table(pro[i])
  colnames(gene)=c("chr", "start", "end", "locus", "chr2", 
                   "start2", "end2", "gene", "dot", "strand", "dist")

  gene=merge(gene, gi, by.x="gene", by.y="gene_id")
  gene$closestGenes=paste(gene$gene, gene$gene_name, gene$dist, sep="|")
  gene=gene[,c("locus", "closestGenes")]
  gene=aggregate(closestGenes~locus, gene, paste, collapse=",")
  ase=merge(ase,gene,by="locus", all.x=T)
  
  gwasCAT$anno<-paste(gwasCAT$study,gwasCAT$LDsnp,gwasCAT$r2_pop.x, sep='|')
  lead<-subset(gwasCAT, !duplicated(gwasCAT[,c("SNP_ID_CURRENT", "study")]))
  lead<-aggregate(study~SNP_ID_CURRENT, lead, paste, collapse=";")
  ld<-subset(gwasCAT, !duplicated(gwasCAT[,c("LDsnp", "anno")]))
  ld<-aggregate(anno~LDsnp, ld, paste, collapse=";")
  
  ase<-merge(ase, lead[, c("SNP_ID_CURRENT", "study")], by.x="snpID", by.y="SNP_ID_CURRENT", all.x=T)
  ase<-merge(ase, ld[, c("LDsnp", "anno")], by.x="snpID", by.y="LDsnp", all.x=T)
  

  
  if (i<3){
  mymot=read.table(motR[i], header=T,stringsAsFactor=F)
  mymot2=subset(mymot, mymot$maxRefScore-mymot$maxAltScore>0)
  
  if (i==1){ase$motDisrHomer<-ase$snpID %in% mymot2[mymot2$Motif=="Nkx2.5_0.93",4 ]}
  if (i==2){ase$motDisrHomer<-ase$snpID %in% mymot2[mymot2$Motif=="CTCF_Boris",4 ]}
  
  mymot[is.na(mymot)]<-0
  mymot=merge(mymot, ase[,c('variantID', 'new_maf')], by.x="SNP", by.y="variantID")
  mymot[mymot$ref_maf>0.5,'scoreRef']<-mymot[mymot$ref_maf>0.5,c('maxRefScore')]
  mymot[mymot$ref_maf<0.5,'scoreRef']<-mymot[mymot$ref_maf<0.5,c('maxAltScore')]
  mymot[mymot$ref_maf>0.5,'scoreAlt']<-mymot[mymot$ref_maf>0.5,c('maxAltScore')]
  mymot[mymot$ref_maf<0.5,'scoreAlt']<-mymot[mymot$ref_maf<0.5,c('maxRefScore')]
  mymot$Diff<-round(mymot$scoreAlt-mymot$scoreRef,2)
  
  mymot$Motif_anno<-paste(mymot$Motif, mymot$Diff,mymot$maxRefScoreStrand, 
                          round(mymot$scoreRef,2), round(mymot$scoreAlt,2), sep="|" )
  
  agg=aggregate(Motif_anno~SNP, mymot, paste, collapse=",")
  ase=merge(ase, agg, by.x="variantID", by.y="SNP", all.x=T)
  #ase=subset(ase, select=-c(anyOthMot, coreMot))
  }
  
  
  if (i>2){
    ase$motDisrHomer<-NA
    ase$Motif_anno<-NA
    #ase$motDisr<-NA
    #ase$scoreDiff<-NA
    ase<-subset(ase, select=colnames(m))
  }

 
  m<-rbind(m,ase)
  
}

m<-subset(m,m$combined_fdr<0.05)

m$topxp<-m$snpID %in% lv$snpID | m$snpID %in% aa$snpID
m$bestp<-m$snpID %in% lv_eqtls$snpID[lv_eqtls$has_best_p==1] | 
  m$snpID %in% aa_eqtls$snpID[aa_eqtls$has_best_p==1]


write.table(m, "Supplementary_table_ase_annotations.txt", sep="\t", row.names=FALSE, quote=FALSE)
```
