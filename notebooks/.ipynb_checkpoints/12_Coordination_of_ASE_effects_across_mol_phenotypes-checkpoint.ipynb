### Coordination of genetic effect across phenotypes :
ChIP-Seq peaks were paired with the closest gene or with ChIP-Seq peaks (from the other datasets) within 500 bp using ‘bedtools closest'. Using linear regression, we tested the association between the family genotypes (0, 1, 2) of the ASE-SNVs (FDR <0.05) and either the corresponding z-score normalized peak coverage or gene expression of the closest feature. Spearman correlation was calculated between ß’s of SNVs that showed ASE in ChIP-Seq datasets (the most significant SNV per peak, FDR< 0.05), and ß’s of the same variant for the closest gene or peak in a different molecular phenotype dataset.

using residuals to correct for batch effects and other variables 

```r
library(stringr)
library(RColorBrewer)
suppressPackageStartupMessages(library(gplots))
```

```r
ase_dir  = 'ASE_chip/correlationWithBeta/'
corr     = paste(ase_dir , "Correlation_ase_snv_", loci, ".txt", sep="")

rna_ipsc = 'PCA_rnaseq/iPSC_29samp/residual_counts.txt'
rna_cms  = 'PCA_rnaseq/iPSC_CM/residual_counts.txt'
chip_tables = paste("PCA_chipseq", loci, 'residual_counts.txt',sep="/")
sample_table = read.csv("PCA_rnaseq/fam1070_data_plus_production-v3.csv", stringsAsFactors = FALSE) 

```

```r
for (c in 1:3) {
ch = read.table(corr[c], header=TRUE, check.names=F)

for (d in 1:5){
if(d!=c) {
if(d<4 ) {

rn              = read.table(chip_tables[d], header=T, check.names=F, row.names=1 )
closest         = loci[d]
samp            = read.table(paste("PeakCalling", loci[d], "sample.use.list", sep="/"),header=F, check.names=F)
colnames(samp)  = c('uuid0', 'Subject_ID', 'UUID')    
    
   
pr               = read.table(paste(home,"Coordination_ASE_effects/matches/",loci[c],"_closest_",loci[d],".bed",sep=""),header=F)
pr$locus         = paste (pr[,1], ":", pr[,2],"-", pr[,3], sep="")
pr$closest_locus = paste (pr[,4], ":", pr[,5],"-", pr[,6], sep="")
colnames(pr)[7]  = "dist"
pr               = subset(pr, pr$dist<500)    

}
 
if( d>3) {
if(d == 5) {  rn       = read.table(rna_ipsc, header=T, check.names=F)
              samp     = sample_table[sample_table$Cell_type == "iPSC" & sample_table$Data_type=="RNA-Seq",]
              closest  = "exp_iPSC"
}

if(d == 4){  rn        = read.table(rna_cms, header=T, check.names=F)
             samp      = sample_table[sample_table$Cell_type == "iPSC-CM"& sample_table$Data_type=="RNA-Seq",]
             closest   = "exp_iPSC_CM"
}
pr                     = read.table(paste (home, "Coordination_ASE_effects/matches/", loci[c], "_closest_gene.bed", sep="" ), header=F)
pr$locus               = paste (pr[,1], ":", pr[,2],"-", pr[,3], sep="")
colnames(pr)[c(7,10)]  = c("closest_locus","dist")
#pr                     =subset(pr, pr$dist<200000)

}

    
    
chp  = merge(ch, pr[,c("locus","closest_locus","dist")], by.x="peakID", by.y="locus" )
chpr = merge(chp [c("closest_locus", "dist","peakID" ,"varID" ,"combined_pv", "combined_fdr" ,"ref_freq", "Coefficient",
             "ID" ,"REF" ,"ALT",'iPSCORE_2_1', 'iPSCORE_2_2', 'iPSCORE_2_3' ,'iPSCORE_2_4', 'iPSCORE_2_6', 'iPSCORE_2_7' ,'iPSCORE_2_9')], 
                rn, by="closest_locus", by.y="row.names")

peaks      = subset(chpr, select=as.character(samp$UUID))        
genotype   = subset(chpr, select=as.character(samp$Subject_ID))
   

res=data.frame()
for( m in 1:nrow(chpr)) {    
    
    df         = data.frame(peaks=t(peaks)[,m], genotype=t(genotype)[,m])
    df$subject = str_split_fixed(colnames(genotype), "\\.", 2)[,1]
    df$peaks   = scale(df$peak) # z-score normalized
    
    mod<-lm(peaks ~ genotype, data=df)  
    cof<-coef(summary(mod))[2,1]
    pv<-coef(summary(mod))[2,4]
    res[m,"Coefficient_closest"]<-cof
    res[m,"lm_pVal_closest"]<-pv
  
  }  


ag = cbind(chpr, res)
write.table(ag, paste("Coordination_ASE_effects/results/Coordination_",  loci[c], "_with_", closest, ".txt", sep=""), 
            sep="\t", row.names=FALSE, col.names=TRUE)

    
    }
}

}
```

```r
rho   = data.frame()
pvals = data.frame()
ns    = data.frame()

pdf("/home/paola/Family1070/private_output/Figures/Scatterplots_red.pdf")
par(mfrow=c(3,3), pin=c(1.3,1.3))
for (c in 1:3) {
for (d in 1:5){
if(d!=c) {
if(d<4 ) {closest = loci[d]}
if(d==4) {closest   = "exp_iPSC_CM"}
if(d==5) {closest   = "exp_iPSC"}    
ag = read.table(paste("Coordination_ASE_effects/results/Coordination_",  loci[c], "_with_", closest, ".txt", sep="")
                 , header=TRUE, check.names=F, sep="\t")
    
ag = ag[order(ag$combined_fdr, decreasing=F),] 
ag = subset(ag, !duplicated(ag$peakID))

colors ="#1B69AF"
colors ='orangered'
xlim = c(-max(abs(ag$Coefficient)),max(abs(ag$Coefficient))) 
ylim = c(-max(abs(ag$Coefficient_closest)),max(abs(ag$Coefficient_closest))) 

plot(Coefficient_closest~Coefficient,data=ag, ylim=ylim, xlim=xlim,
     pch=16, cex=0.5, col=colors, 
     cex.lab=1.1, cex.axis=1.1, main= paste(loci[c], closest, sep= "-"), xlab = paste("Effect size", loci[c]),
     ylab = paste("Effect size", closest))

abline(h=0,v=0, lty=2, col='gray30', lwd=2)

  
co = cor.test(ag$Coefficient, ag$Coefficient_closest, method="spearman", exact=F)
l  = lm(Coefficient_closest~Coefficient,data=ag)
abline(l)
text (0.8, -1.5, paste( "Rho=",round(co$estimate,2), "\n p.val=",signif(co$p.value,3) ),cex=1.1)

rho   [c,d]<- co$estimate
pvals [c,d]<- co$p.value
ns    [c,d]<- nrow(ag)

        }
else {   
     rho   [c,d] <- NA
     pvals [c,d] <- NA
     ns    [c,d] <- NA
      }
    }
}
dev.off()
rownames(rho) = rownames(pvals) = rownames(ns) = loci
colnames(rho) = colnames(pvals) = colnames(ns) = c(loci, "exp.CMs", "exp.iPSCs")
```
