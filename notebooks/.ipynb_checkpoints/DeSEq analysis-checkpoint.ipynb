```r
ta.sub=round(ta.sub,0)
dds<-DESeqDataSetFromMatrix(ta.sub, group, design= ~ cTNT_TPM+Cell_type+subject) 
dds<-DESeq(dds, fitType="parametric")

subjects=c("S07001", "S07002", "S07003", "S07004", "S07006", "S07007", "S07009")
FDR=0.05

for (i in 1:6) { 
  for (j in (i+1):7) { 
  	s1=subjects[i]
    s2=subjects[j]
    pair=paste(s1, "vs", s2, sep="")
    cat(pair, "\n")

res<-results(dds, contrast=c("subject", s1, s2))
summ=sum(res$padj < FDR, na.rm=TRUE)

res<-as.data.frame(res)
write.table(res,file=paste(pair,".txt", sep="" ),sep="\t", row.names=TRUE, col.names=NA)
columnames=c( "", paste (pair, c("log2FC","PValue","FDR"), sep="."))
 
 if ((i ==1)&(j==2)){
      fc = data.frame(res[,2], row.names = row.names(res))
      colnames(fc)= columnames[2]
      pval = data.frame(res[,5], row.names = row.names(res))
      colnames(pval)= columnames[3]
      fdr = data.frame(res[,6], row.names = row.names(res))
      colnames(fdr)= columnames[4]
      finalsum= summ
      pair.list=pair
      
      
    } else {
      fc = cbind(fc, res[,2])
      colnames(fc)[[length(colnames(fc))]] = columnames[2]
	  pval = cbind(pval, res[,5])
      colnames(pval)[[length(colnames(pval))]] = columnames[3]
      fdr = cbind(fdr, res[,6])
      colnames(fdr)[[length(colnames(fdr))]] = columnames[4]
      finalsum=c(finalsum,summ)  
      pair.list=c(pair.list,pair)
     

    }
}
}

write.table(fc, file="FC_summary.txt", sep="\t", row.names=TRUE, col.names=NA)
write.table(pval, file="Pval_summary.txt", sep="\t", row.names=TRUE, col.names=NA)
write.table(fdr, file="FDR_summary.txt", sep="\t", row.names=TRUE, col.names=NA)

diP=finalsum
diP.f=diP/nrow(res)

```

## Enrichment using logistic regression for differential SNPs

```r
gdi=read.csv("/home/paola/SNP_intersection/fam1070_prom_by_gene.rs.alleles.csv")
### remove indels
gdi.f=subset(gdi, (gdi$REF=="A"|gdi$REF=="T"|gdi$REF=="C"|gdi$REF=="G"|gdi$REF=="*"))
gdi.f=subset(gdi.f, (gdi.f$ALT=="A"|gdi.f$ALT=="T"|gdi.f$ALT=="C"|gdi.f$ALT=="G"|gdi.f$ALT=="*"))
gdi<-gdi.f
#############################
# PROMOTER DEFINITION= -2000,+200 BP FORM TSS OF EACH TRANSCRIPTS, THEN MERGED BY GENE (Chris' approach)
prom<- read.table('/publicdata/gencode_v19_20151104/promoters_by_gene.bed')

colnames(prom)<-c("chrom", "start", "end", "promoter", "strand")
d<-str_split_fixed(prom$promoter,"\\_",2)
prom$geneID<-d[,1]
prom$size<-prom$end-prom$start

ag=aggregate(size~geneID, prom, sum)
```

```r
stats=data.frame()
c=1
gi=gene_info

fc=read.table("FC_summary.txt", header=T)
fc=fc[order(fc[,1]),]

pv=read.table("FDR_summary.txt", header=T)
pv=pv[order(pv[,1]),]


colnames(fc)[1]<-"geneID"
int=merge(fc[,1:2], gdi.f, by.x="geneID", by.y="gene", all.x=TRUE )

int<-int[,c(1,22:42)]
int[,2:22][is.na(int[,2:22])] <- 0   

int<-merge(int, ag, by.x=1, by.y=1)
head(int)
sizes=int[!duplicated(int$geneID),c(1,23)]
    
######## Table1 with real differences
gen_dist<-matrix(NA, ncol=21, nrow=nrow(fc))

for (i in 1:21){ 
  p=aggregate(int[,i+1], list(gene=int$geneID), sum)
  gen_dist[,i]=p[,2]
}

gen_dist<-(gen_dist/sizes$size)*1000
gen_dist[is.na(gen_dist)]<-0
colnames(gen_dist)=colnames(int)[2:22]
rownames(gen_dist)<-as.character(sizes$geneID)

gen_dist=gen_dist[order(rownames(gen_dist)),]

######## Table with absolute differences
int[, 2:22]<- abs(int[,2:22])

abs_gen_dist<-matrix(NA, ncol=21, nrow=nrow(fc))

for (i in 1:21){ 
  p=aggregate(int[,i+1], list(gene=int$geneID), sum)
  abs_gen_dist[,i]=p[,2]
}

abs_gen_dist<-(abs_gen_dist/sizes$size)*1000
abs_gen_dist[is.na(abs_gen_dist)]<-0
colnames(abs_gen_dist)=colnames(int)[2:22]
rownames(abs_gen_dist)<-as.character(sizes$geneID)

abs_gen_dist=abs_gen_dist[order(rownames(abs_gen_dist)),]


dt<-data.frame()
  
  for (i in 2:22) {
    
    m=data.frame(fc[,1], abs(fc[,i]), pv[,i], abs_gen_dist[,i-1])
    colnames(m)=c("peakID", "absLogFC","FDR","Genetic_distance")
    m$pair<- as.factor(colnames(gen_dist)[i-1])
    m$normFC<-2^(m$absLogFC-mean(m$absLogFC[m$Genetic_distance==0], na.rm=T))
    dt<-rbind(dt, m)
    
  }
  abs_vals<-dt
    
       # remove HLA genes and pseudogenes

abs_vals<-merge(abs_vals, gi[,c(1:3)], by=1)
abs_vals=abs_vals[grep("HLA", as.character(abs_vals$gene_name), invert=T), ]
abs_vals=abs_vals[abs_vals$gene_type!="pseudogene",] 

write.table(abs_vals, "pairwise_absFC.txt", row.names=FALSE, sep="\t", col.names=T)

    
  dt<-data.frame()
  
  for (i in 2:22) {
    
    m=data.frame(fc[,1], fc[,i], pv[,i], gen_dist[,i-1], abs_gen_dist[,i-1])
    colnames(m)=c("peakID", "LogFC","FDR","directional_Genetic_distance", "Genetic_distance")
    m$pair<- as.factor(colnames(gen_dist)[i-1])
    m$normFC<-2^(abs(m$LogFC)-mean(m$LogFC[m$Genetic_distance==0], na.rm=T))
    dt<-rbind(dt, m)
    
  }
     
    
    # remove HLA genes and pseudogenes

dt<-merge(dt, gi[,c(1:3)], by=1)
dt=dt[grep("HLA", as.character(dt$gene_name), invert=T), ]
dt=dt[dt$gene_type!="pseudogene",]


ref<-as.data.frame(subset(dt, dt$LogFC*dt$directional_Genetic_distance<=0 ))
write.table(ref, "pairwise_ref_pref.txt", row.names=FALSE, sep="\t", col.names=T)

alt<-as.data.frame(subset(dt, dt$LogFC*dt$directional_Genetic_distance>=0 ))
write.table(alt, "pairwise_alt_pref.txt", row.names=FALSE, sep="\t", col.names=T)
   
      ### Linear regression  ### Linear regression  ### Linear regression  ### Linear regression
    stats<-matrix(NA, ncol=3, nrow=6)
  colnames(stats)<-c("Absolute", "Ref.", "ALt.")
  rownames(stats)<-c("B_lm", "P_lm", "B_lm_norm", "P_lm_norm", "B_glm", "P_glm")
  
  ### Linear regression
  l=lm(absLogFC~Genetic_distance+pair, data=abs_vals)
  stats[1,1]<-l_slope<- summary(l)$coefficients[2,1]
  stats[2,1]<-summary(l)$coefficients[2,4]
  
  l=lm(abs(LogFC)~Genetic_distance+pair, data=ref)
  stats[1,2]<-summary(l)$coefficients[2,1]
  stats[2,2]<-summary(l)$coefficients[2,4]
  
  l=lm(abs(LogFC)~Genetic_distance+pair, data=alt)
  stats[1,3]<-summary(l)$coefficients[2,1]
  stats[2,3]<-summary(l)$coefficients[2,4]
  
  ### Linear regression on normalized values
  l=lm(normFC~Genetic_distance+pair, data=abs_vals)
  stats[3,1]<-l_slope<- summary(l)$coefficients[2,1]
  stats[4,1]<-summary(l)$coefficients[2,4]
  
  l=lm(normFC~Genetic_distance+pair, data=ref)
  stats[3,2]<-summary(l)$coefficients[2,1]
  stats[4,2]<-summary(l)$coefficients[2,4]
  
  l=lm(normFC~Genetic_distance+pair, data=alt)
  stats[3,3]<-summary(l)$coefficients[2,1]
  stats[4,3]<-summary(l)$coefficients[2,4]
  
  ### Logistic regression
  gl=glm(FDR<0.05~Genetic_distance+pair, data=abs_vals, binomial)
  stats[5,1]<-gl_slope<-summary(gl)$coefficients[2,1]
  stats[6,1]<-summary(gl)$coefficients[2,4]
  
  gl=glm(FDR<0.05~Genetic_distance+pair, data=ref,  binomial)
  stats[5,2]<-summary(gl)$coefficients[2,1]
  stats[6,2]<-summary(gl)$coefficients[2,4]
  
  gl=glm(FDR<0.05~Genetic_distance+pair, data=alt, binomial)
  stats[5,3]<-summary(gl)$coefficients[2,1]
  stats[6,3]<-summary(gl)$coefficients[2,4]
  
  write.csv(stats, "Summary_regression_statistics.csv")
    

```

```r
color="red"
## ## ## ## ## ## ## ## ## ## ## ## 
## Bin tables for plotting  ## ## ## 

#pdf("Aggregate_Results.pdf")
par(mfrow=c(3,2))
li=list(abs_vals, ref, alt) #, null)
for_plot<-data.frame(c(0:6))

for (j in 1:length(li)){
m<-li[[j]]
m$Genetic_distance<-round(m$Genetic_distance,0)
m[m$Genetic_distance>5,'Genetic_distance'] <-6

    
binned<-aggregate(FDR<0.05~Genetic_distance, m, mean, na.rm=T)
plot(binned, bg=color, type="l", lwd=2)
binnedfc<-aggregate(normFC~Genetic_distance, m, mean, na.rm=T)
plot(binnedfc, bg=color, type="l", lwd=2)
    
for_plot<-data.frame(for_plot, binned[,2], binnedfc[,2])
    
}
#dev.off()
colnames(for_plot)<-c("variation_rate",'Prop_differential', 'absolute_fold_change','Prop_differential_ref', 'absolute_fold_change_ref',
                      'Prop_differential_alt', 'absolute_fold_change_alt')

write.table(for_plot, "Binned_averages_to_plot.txt", sep="\t", col.names=T)

```
