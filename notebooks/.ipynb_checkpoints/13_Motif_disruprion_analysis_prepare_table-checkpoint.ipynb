```r
ase<-read.table(paste(name,"TestMAtr.txt",sep="/"), header=T)
mo<-read.table(paste(name,'analysis/All_siteScores_1.txt',sep="/"), header=T, stringsAsFactors=F)
mo$motif_class = substr(mo$Motif, 1, 3)
mo$motif_class = toupper(mo$motif_class)
mo$motif_class[mo$Motif=="Tbx5"]<-"Tbx5"  ### Tbx5 and 20 are different
mo$disrupt = (mo$maxRefScore >= mo$TFM_score|mo$maxAltScore >= mo$TFM_score) & abs(mo$maxRefScore - mo$maxAltScore) > 0
mot_table<-merge( mo, ase, by.x="SNP", by.y="Variant" )
```

```r
#pdf(paste(name,"analysis/EnrichmentResults1.pdf",sep="/"))
pdf("/frazer01/home/paola/Family1070/private_output/Figures/Motif_disruption_enrichment.pdf")
par(pin=c(2.5,1))

for (l in 1:2){
    cat(l)
#Separate analysis per motif
    if(l==1) {
        motifs<-levels(as.factor(mot_table$Motif))
        mo<-split(mot_table, as.factor(mot_table$Motif))
    }  

#Combine per motif class
    if(l==2) {
          motifs<-levels(as.factor(mot_table$motif_class))
          mot_table2 = mot_table[order(abs(mot_table$maxRefScore - mot_table$maxAltScore),decreasing=TRUE),]
          mot_table2 = mot_table2[!duplicated(mot_table2[,c("motif_class", "SNP")]),]
          mo = split(mot_table2, as.factor(mot_table2$motif_class))
    }
    


res1<-data.frame() # Fisher test x=ASE/NoASE y=NoMotif/MotifCHange 

for (m in 1:length(motifs)){
  test<-mo[[m]]
  tab1<-table(test$combined_fdr<0.05, test$disrupt==TRUE)
  fi1<-fisher.test(tab1)
  res1[1:6,m]<-c(tab1, round(fi1$estimate,2), fi1$p.value)
  
  
}

colnames(res1)<- motifs
rownames(res1)<- c("NoASE-NoMotif", "ASE-NoMotif", "NoASE-Motif", "ASE-Motif", "OddsRatio", "P.value")
res1["FDR-p.value",]<-p.adjust(res1["P.value",], method="BH")

write.table(res1,paste(name,"/analysis/EnrichmentResults",l, ".txt",sep=""), row.names=T, col.names=T , sep="\t", quote=F)


my_palette <- colorRampPalette(c("white","orangered3" ))(n = 20)
#my_palette =colorRampPalette( rev(brewer.pal(9, "Reds")) )(20)[21:1] 

res1<-res1[,order(res1["P.value",], decreasing=F)]
x<-as.numeric((-log(res1["FDR-p.value",],10)))

x[x<1.3]<-NA    
color<-my_palette[as.numeric(cut(x,breaks = 20))]
par(pin=c(2.5,1))
barplot(as.numeric(res1["OddsRatio",]), names.arg=colnames(res1), 
        las=2, ylim=c(0, max(res1["OddsRatio",])+0.5), 
        ylab="Odds ratio", space=c(0.2,0.2), col=color)
abline(h=1, col="red", lwd=2, lty=2)
text((1:length(motifs)), res1["OddsRatio",]+0.3, c( "","*") [(res1["FDR-p.value",]<0.05)+1], cex=2)
colkey(my_palette, clim=range(x, na.rm=T), clab="-log10(FDR)", add=T, side=4, length=0.5)
    
}

dev.off()
```

### Frequency of motif alteration and direction with respect to ASE

```r
sigSites<-read.table(paste(name,'analysis/AlleleAlteringSites_scores_1.txt',sep="/"), header=T, stringsAsFactors=F)
sigSites<-merge( sigSites, ase, by.x="SNP", by.y="Variant" )
sigSites= sigSites[sigSites$combined_fdr<0.05,]
dim(sigSites)
```

```r
sigSites$Score_change = sigSites$maxRefScore-sigSites$maxAltScore
```

```r
sigSites$Motif <- factor(sigSites$Motif)
sigSites$same_effect = (sigSites$ref_freq > 0.5 & sigSites$Score_change >0 ) | (sigSites$ref_freq < 0.5 & sigSites$Score_change < 0)
```

```r
up=NA
down=NA
for (f in 1:length(levels(sigSites$Motif))) {
  mo=sigSites[sigSites$Motif==levels(sigSites$Motif)[f],]
  up[f]  <- sum(mo$same_effect)
  down[f]<- sum(mo$same_effect==FALSE)
}

tab=rbind(up,down)
colnames(tab)=levels(sigSites$Motif)
write.csv(tab, paste(name,"analysis/Table_for barplot.csv",sep="/"))
```

### Combine motifs from the same class
Althougth there are more sites with NKX2.1 binding than of NKX2-5, they have less correlation with ASE

```r
sigSites$motif_class= substr(sigSites$Motif, 1, 3)
sigSites$motif_class = toupper(sigSites$motif_class)
```

```r
aggregateMotifs = function(motif_class) {
categ = sigSites[sigSites$motif_class == motif_class,]
categ = categ[order(categ$Motif),]
ag = aggregate(Motif~SNP ,categ, paste, collapse=",")
ag2 = aggregate(Motif~SNP ,categ, length)
ag3 = aggregate(Score_change~SNP ,categ, mean)
ag4 = aggregate(ref_freq~SNP ,categ, mean)
agg = cbind(ag,ag2[,2],ag3[,2],ag4[,2] )
colnames(agg) = c('SNP',"Motifs", "Number_motifs", "Score_change","ref_freq" )
agg$Motifs = as.factor(agg$Motifs)
agg = agg[order(agg$Number_motifs),]
    
    return(agg)
    }
```

```r
agg = aggregateMotifs ("NKX")
```

```r
colors = c("black", "gray", "orangered2", "green", "orange", "yellow", "dodgerblue3","forestgreen")

par(mfrow=c(2,2), pin=c(2.2,2.2))
plot(Score_change~ ref_freq, agg,  pch = 20, cex = agg$Number_motifs, 
     col= colors[agg$Motifs])
  abline(h=0,v=0.5, col="gray")

for (f in 1:length(levels(agg$Motifs))) {
    
mo = agg[agg$Motifs==levels(agg$Motifs)[f],]
  l  = lm(Score_change~ ref_freq, mo)
if (!is.na(summary(l)[[4]][2]) ) {
    abline(l, col =colors[f], lwd = 2, lty=2) 
}
    }
plot.new()
legend("center",legend= unique(agg$Motifs), col= colors[unique(agg$Motifs)],  pch = 20, pt.cex=c(1,1,1,2,2,2,3) )
hist( agg$Number_motifs)
```

```r
motifs_to_aggregate = c("NKX", "GAT", "TEA", "MEF")

for (i in 1:length(motifs_to_aggregate)){
agg = aggregateMotifs (motifs_to_aggregate [i])
    agg$Motif = motifs_to_aggregate [i]
if (i ==1) { agg_tab = agg } 
    else   { agg_tab = rbind (agg_tab, agg)}
}
```

```r
agg_tab$same_effect = (agg_tab$ref_freq > 0.5 & agg_tab$Score_change >0 ) | (agg_tab$ref_freq < 0.5 & agg_tab$Score_change < 0)
```

```r
sigSites_cut = droplevels(subset (sigSites, !(motif_class %in% motifs_to_aggregate)))
```

```r
sigSites_new = rbind (sigSites_cut[,c('SNP','ref_freq','Motif','same_effect')], agg_tab[,c('SNP','ref_freq','Motif','same_effect')])
```
